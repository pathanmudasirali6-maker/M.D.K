```python
import pandas as pd

df = pd.read_csv('/content/matches.csv')
print("DataFrame 'df' loaded successfully. Displaying the first 5 rows:")
print(df.head())
```

```python
print("\n--- DataFrame Information ---\n")
df.info()

print("\n--- Descriptive Statistics for Numerical Columns ---\n")
print(df.describe())

print("\n--- Descriptive Statistics for All Columns ---\n")
print(df.describe(include='all'))
```

```python
import matplotlib.pyplot as plt

# Calculate the number of missing values for each column
missing_values_count = df.isnull().sum()

# Calculate the percentage of missing values for each column
missing_values_percentage = (missing_values_count / len(df)) * 100

# Filter out columns with no missing values for better visualization
missing_values_percentage = missing_values_percentage[missing_values_percentage > 0].sort_values(ascending=False)

print("--- Percentage of Missing Values per Column ---")
print(missing_values_percentage)

# Create a bar plot for the missing value percentages
if not missing_values_percentage.empty:
    plt.figure(figsize=(12, 6))
    missing_values_percentage.plot(kind='bar', color='skyblue')
    plt.xlabel('Columns', fontsize=12)
    plt.ylabel('Percentage of Missing Values', fontsize=12)
    plt.title('Percentage of Missing Values per Column', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found in the dataset.")
```

```python
import matplotlib.pyplot as plt

# Calculate the value counts for the 'result' column
result_counts = df['result'].value_counts()

print("--- Distribution of Match Results ---")
print(result_counts)

# Create a bar plot for the distribution of match results
plt.figure(figsize=(8, 6))
result_counts.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Match Results', fontsize=14)
plt.xlabel('Match Result', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the value counts for the 'winner' column
winner_counts = df['winner'].value_counts()

print("\n--- Distribution of Winning Teams ---")
print(winner_counts)

# Get the top 10 most frequent winning teams
top_10_winners = winner_counts.head(10)

# Create a bar plot for the top 10 most frequent winning teams
plt.figure(figsize=(12, 7))
top_10_winners.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Most Frequent Winning Teams', fontsize=14)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Number of Wins', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Histogram for 'win_by_runs'
plt.figure(figsize=(10, 6))
df['win_by_runs'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Wins by Runs', fontsize=14)
plt.xlabel('Runs (Winning Margin)', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()

# 2. Box plot for 'win_by_runs'
plt.figure(figsize=(10, 6))
df['win_by_runs'].plot(kind='box', vert=False, patch_artist=True, boxprops=dict(facecolor='lightcoral'))
plt.title('Box Plot of Wins by Runs', fontsize=14)
plt.xlabel('Runs (Winning Margin)', fontsize=12)
plt.ylabel(' ', fontsize=12) # Empty label to remove 'win_by_runs'
plt.tight_layout()
plt.show()

# 3. Histogram for 'win_by_wickets'
plt.figure(figsize=(10, 6))
df['win_by_wickets'].plot(kind='hist', bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Wins by Wickets', fontsize=14)
plt.xlabel('Wickets (Winning Margin)', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()

# 4. Box plot for 'win_by_wickets'
plt.figure(figsize=(10, 6))
df['win_by_wickets'].plot(kind='box', vert=False, patch_artist=True, boxprops=dict(facecolor='lightsalmon'))
plt.title('Box Plot of Wins by Wickets', fontsize=14)
plt.xlabel('Wickets (Winning Margin)', fontsize=12)
plt.ylabel(' ', fontsize=12) # Empty label to remove 'win_by_wickets'
plt.tight_layout()
plt.show()

# 5. Bar plot for 'season'
plt.figure(figsize=(10, 6))
df['season'].value_counts().sort_index().plot(kind='bar', color='lightgray', edgecolor='black')
plt.title('Number of Matches Per Season', fontsize=14)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
